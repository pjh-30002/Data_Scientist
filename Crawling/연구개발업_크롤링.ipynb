```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("http://www.rndservice.or.kr/info/regist.do")
driver.implicitly_wait(20)
Last_Com = []
Last_CEO = []
Last_tel = []
Last_no = []
Last_main = []
Last_num = []
Last_money = []
Last_sort = []
C_url = []

try:
    for k in range(1,14):
        for j in range(3,13):
            for i in range(1,11):
                #업체명
                Company_name = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[1]/strong".format(i)).text
                #대표자
                C_CEO = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[1]/div[2]".format(i)).text
                # 연락처
                C_tel = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[2]/div[2]".format(i)).text
                # 신고번호
                C_no = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[3]/div[2]".format(i)).text
                # 주요사업
                C_main = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[4]/div[2]".format(i)).text
                # 종사자수
                C_num = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[1]/div[4]".format(i)).text
                # 매출액
                C_money = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[2]/div[4]".format(i)).text
                # 신고업종
                C_sort = driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[2]/div[{}]/div[2]/div/div[3]/div[4]".format(i)).text
                
                Last_Com.append(Company_name)
                Last_CEO.append(C_CEO)
                Last_tel.append(C_tel)
                Last_no.append(C_no)
                Last_main.append(C_main)
                Last_num.append(C_num)
                Last_money.append(C_money)
                Last_sort.append(C_sort)
                
                # Url
                a_tags = soup.find_all("a", {'class':"sub6_1__item_link"})

                for a_tag in a_tags :
                    C_url.append(a_tag)
                print(Company_name)
                # 다음페이지로    
            driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[3]/div/button[{}]".format(j)).click()
            driver.implicitly_wait(10)
    print("{} 번째 장을 마칩니다".format(k))
except:
    print("끝")
```

```python
df = pd.DataFrame()
df['업종명'] = Last_Com
df['대표자명'] = Last_CEO
df['전화번호'] = Last_tel
df['신고번호'] = Last_no
df['주요사업'] = Last_main
df['종사자수'] = Last_num
df['매출액'] = Last_money
df['신고업종'] = Last_sort

df
```

```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("http://www.rndservice.or.kr/info/regist.do")
driver.implicitly_wait(20)
C_url_l = []

try:
    for k in range(1,21):
        time.sleep(2)
        for j in range(3,13):
            time.sleep(2)
            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")
            # Url
            a_tags = soup.find_all("a", {'class':"sub6_1__item_link"})
            for a_tag in a_tags :
                C_url_l.append(a_tag['href'])
                print("-",a_tag['href'])
            time.sleep(2)    
                # 다음페이지로    
            driver.find_element_by_xpath("/html/body/div/div/div/div/div/form/div[3]/div/button[{}]".format(j)).click()
            driver.implicitly_wait(10)
    print("{} 번째 장을 마칩니다".format(k))
except:
    print("끝")
```
