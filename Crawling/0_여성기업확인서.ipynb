```python
import pandas as pd
import numpy as np
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.common.keys import Keys
pd.set_option('display.max_row', 20)
pd.set_option('display.max_columns', 100)
```

```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("http://www.wbiz.or.kr/web/entsearch/BD_index.do")
driver.implicitly_wait(20)
driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/div[2]/form/fieldset/select/option[8]").click()
driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/div[2]/form/fieldset/img").click()
time.sleep(2)

a = []

try:
    for k in range(1,200):
        if k == 1:
            for j in range(2,12):
                if j ==2 :
                    
                    html = driver.page_source
                    soup = BeautifulSoup(html,"html.parser")
                    needs = soup.find_all("tr")

                    for i in needs:
                        need = i.text.split('\n')
                        a.append(need)
                    driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/div[2]/div[2]/div/a[{}]".format(j)).click()
                    driver.implicitly_wait(20)
                else:
                    html = driver.page_source
                    soup = BeautifulSoup(html,"html.parser")
                    needs = soup.find_all("tr")

                    for i in needs[1:]:
                        need = i.text.split('\n')
                        a.append(need)
                    driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/div[2]/div[2]/div/a[{}]".format(j)).click()
                    driver.implicitly_wait(20)
        else:
            for j in range(4,14):
                time.sleep(1)
                html = driver.page_source
                soup = BeautifulSoup(html,"html.parser")
                needs = soup.find_all("tr")
                
                for i in needs[1:]:
                    need = i.text.split('\n')
                    a.append(need)
                driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/div[2]/div[2]/div/a[{}]".format(j)).click()
                driver.implicitly_wait(20)
                    
        print(k)
except:
    print("종료")
    
col1 = []
col2 = []
col3 = []
col4 = []
col5 = []
col6 = []

col = ['번호','기업명','대표자','업종/업태','주소','취급품목']
print(col)
for n in range(1,len(a)):
    col1.append(a[n][1])
    col2.append(a[n][2])
    col3.append(a[n][3])
    col4.append(a[n][4])
    col5.append(a[n][5])
    col6.append(a[n][6])

    
df = pd.DataFrame(columns = col)
df['번호'] = col1
df['기업명'] = col2
df['대표자'] = col3
df['업종/업태'] = col4
df['주소'] = col5
df['취급품목'] = col6
df.to_excel(r"C:/Users/KED_Internet/바탕화면/ESG/monthly_data/ESG_여성기업확인서_{}_{}_{}.xlsx".
            format(time.localtime()[0],time.localtime()[1],time.localtime()[2]), index=None)
```
