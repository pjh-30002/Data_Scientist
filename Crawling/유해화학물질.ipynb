```python
import pandas as pd
import numpy as np
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
pd.set_option('display.max_row', 20)
pd.set_option('display.max_columns', 100)
```

```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("https://icis.me.go.kr/prtr/prtrInfo/entrpsSearch.do")
driver.implicitly_wait(20)
driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[3]/div[1]/div[1]/dl[1]/dd/select/option[3]").click()
driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[3]/div[2]/a[1]").click()
# 100개씩
driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[4]/div[2]/select/option[4]").click()
# 클릭
driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[4]/div[2]/a").click()
driver.implicitly_wait(20)

names=[]
addresses=[]
outs =[]
puts = []
moves=[]
Counts = 1

try:
    for k in range(1,5):
        for j in range(3,13):
            for i in range(1,101):
                # 업체명
                name = driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[7]/table/tbody/tr[{}]/td[2]/a".format(i)).text
                # 주소
                address = driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[7]/table/tbody/tr[{}]/td[3]".format(i)).text
                # 배출량
                out = driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[7]/table/tbody/tr[{}]/td[4]".format(i)).text
                # 자가매립량
                put = driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[7]/table/tbody/tr[{}]/td[5]".format(i)).text
                # 이동량
                move = driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[7]/table/tbody/tr[{}]/td[6]".format(i)).text

                names.append(name)
                addresses.append(address)
                outs.append(out)
                puts.append(put)
                moves.append(move)

                print(Counts,".",name,address,out,put,move)
                Counts +=1
            driver.find_element_by_xpath("/html/body/div[2]/div/form[1]/div/div[8]/a[{}]".format(j)).click()
        print(k)
except:
    print("종료")
```
