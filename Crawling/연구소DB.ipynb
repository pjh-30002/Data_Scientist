```python
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
```

```python
df = pd.read_excel("./연구소_v02.xlsx")
df
```

```python
df = df.rename(columns = {'d':'col1'})
```

```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("https://www.rnd.or.kr/user/infoservice/search5.do")
driver.implicitly_wait(20)

aa = []
for i in range(len(df['col1'])):
    try:
        a1 = df['col1'][i][:3]
        a2 = df['col1'][i][4:6]
        a3 = df['col1'][i][7:12]
    except:
        aa.append("")
        continue
    
    elem1 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[1]")
    elem2 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[2]")
    elem3 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[3]")
    
    elem1.send_keys(a1)
    elem2.send_keys(a2)
    elem3.send_keys(a3)
    driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[2]/button[2]").click()
    
    try:
        name = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[3]/div/div[2]/table/tbody/tr/td[2]/span/a").text
        aa.append(name)
    except:
        aa.append("")
    
    elem1 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[1]")
    elem2 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[2]")
    elem3 = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/div[2]/form/div/div[1]/table/tbody/tr[2]/td/input[3]")
    
    elem1.clear()
    elem2.clear()
    elem3.clear()
    time.sleep(1)
```

```python
kk = pd.DataFrame()
kk['이름'] = aa
kk.to_excel('./연구소검증_210708.xlsx')
```
