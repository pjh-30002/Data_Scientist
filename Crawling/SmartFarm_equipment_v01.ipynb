```python
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
pd.set_option('display.max_row', 20)
pd.set_option('display.max_columns', 100)
```

```python
driver = webdriver.Chrome(executable_path= r'C:/Chromedrive/chromedriver')
driver.get("https://www.smartfarmkorea.net/company/equipmentInfo.do?menuId=M01020211")
driver.implicitly_wait(10)

title_L = []
person_L = []
zip_L = []
add_road_L =[]
add_L = []
count=1
try:
    for j in range(1,10):
        for i in range(1,11):
            driver.find_element_by_xpath("/html/body/div/div[4]/div/div[2]/div[2]/form[1]/div[1]/table/tbody/tr[{}]/td[9]".format(i)).click()
            driver.implicitly_wait(20)
            time.sleep(1)
            driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/div/ul/li[6]/a").click()

            title = driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/form[6]/div[1]/table/tbody/tr[1]/td[1]").text
            person = driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/form[6]/div[1]/table/tbody/tr[2]/td").text
            title_L.append(title)
            person_L.append(person)
            print("{}.{}/{}".format(count,title,person))
            count += 1
            # html 파싱
            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")

            # 우편번호
            zip_code_tags = soup.find_all("input", {'id':"cpZipcode"})
            # 도로명주소
            zip_cpaddrRoads = soup.find_all("input", {'id':"cpAddrRoad"})
            # 지번주소
            zip_cpaddrs = soup.find_all("input", {'id':"cpAddr"})

            for zip_code_tag in zip_code_tags:
                zip_L.append(zip_code_tag['value'])

            for zip_cpaddrRoad in zip_cpaddrRoads:
                add_road_L.append(zip_cpaddrRoad['value'])

            for zip_cpaddr in zip_cpaddrs:
                add_L.append(zip_cpaddr['value'])
                print(zip_cpaddr['value'])
            time.sleep(0.5)
            try:
                driver.back()

            except:
                driver.back()

        try:
            driver.find_element_by_xpath("/html/body/div/div[4]/div/div[2]/div[2]/form[1]/div[2]/span[1]/a[{}]".format(j)).click()
            driver.implicitly_wait(10)
            time.sleep(1)
        except:
            driver.find_element_by_xpath("/html/body/div/div[4]/div/div[2]/div[2]/form[1]/div[2]/span[2]/a[{}]".format(j)).click()
            driver.implicitly_wait(10)
            time.sleep(1)                
except:
    print("다음으로")

for i in range(1,11):
    driver.find_element_by_xpath("/html/body/div/div[4]/div/div[2]/div[2]/form[1]/div[1]/table/tbody/tr[{}]/td[9]".format(i)).click()
    driver.implicitly_wait(20)
    time.sleep(1)
    driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/div/ul/li[6]/a").click()

    title = driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/form[6]/div[1]/table/tbody/tr[1]/td[1]").text
    person = driver.find_element_by_xpath("/html/body/div[1]/div[4]/div/div[2]/div[2]/div[2]/form[6]/div[1]/table/tbody/tr[2]/td").text
    title_L.append(title)
    person_L.append(person)
    print("{}.{}/{}".format(count,title,person))
    count += 1
    # html 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # 우편번호
    zip_code_tags = soup.find_all("input", {'id':"cpZipcode"})
    # 도로명주소
    zip_cpaddrRoads = soup.find_all("input", {'id':"cpAddrRoad"})
    # 지번주소
    zip_cpaddrs = soup.find_all("input", {'id':"cpAddr"})

    for zip_code_tag in zip_code_tags:
        zip_L.append(zip_code_tag['value'])

    for zip_cpaddrRoad in zip_cpaddrRoads:
        add_road_L.append(zip_cpaddrRoad['value'])

    for zip_cpaddr in zip_cpaddrs:
        add_L.append(zip_cpaddr['value'])
        print(zip_cpaddr['value'])
    time.sleep(0.5)
    try:
        driver.back()

    except:
        driver.back()
```
